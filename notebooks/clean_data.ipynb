```python
import pandas as pd
```

```python
zomato = pd.read_csv("../raw/zomato.csv", encoding="latin1")
zomato.head()
```

```python
countries = pd.read_excel("../raw/Country-Code.xlsx")
countries.head()
```

```python
# merge files zomato and countries by column "CountryCode"

df = pd.merge(zomato, countries, on="Country Code", how="left")
df.head() 
```

```python
# Rows, Columns
print(df.shape)

# Data types + nulls
df.info()

# Statistical
df.describe()
```

```python
# kis column me kitne null values hain 
df.isnull().sum()

# Drop rows from null cuisines
df = df.dropna(subset=["Cuisines"]) 

# Remove unnecessary columns from dataset

df.drop(columns=["Restaurant ID", "Country Code", "Address", "Locality Verbose", "Longitude", "Latitude", "Is delivering now",
                 "Switch to order menu", "Rating color"], axis=1, inplace=True)
print(df.info())
```

```python
# Remove duplicates
df.duplicated().sum()
df.drop_duplicates(inplace=True)

df.drop_duplicates(subset=["Restaurant Name", "City"], inplace=True)
print("Rows after removing duplicates", df.shape[0])
```

```python
# Change columns names into simplest form

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(df.info())
```

```python
# filtering rows 

df = df[df["aggregate_rating"] != 0]
```

```python
# Reset Indexing

df.reset_index(drop=True, inplace=True)
```
